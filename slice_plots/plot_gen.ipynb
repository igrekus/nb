```python
import itertools

import pandas as pd
import seaborn as sb

from numpy import nan
```

```python
df = pd.read_excel(r'.\data\master_full.xlsx', engine='openpyxl')
df_f = pd.read_excel(r'.\data\master_filtered_full.xlsx', engine='openpyxl')
df_f.head()
```

```python
df.replace([nan], '', inplace=True)
df_f.replace([nan], '', inplace=True)
df_f.head()
```

```python
mats = sorted(filter(bool, set(df['mat'])))
glues = sorted(filter(bool, set(df['glue'])))
glue_paper = sorted(filter(bool, set(df['glue_paper'])))
rate = sorted(filter(bool, set(df['rate'])))
pattern = sorted((filter(bool, set(df['pattern']))))
cell_size = sorted((filter(bool, set(df['cell_size']))))
strip = sorted((filter(bool, set(df['strip']))))
glue_paper
```

```python
mat_vs_glue = list(itertools.product(mats, glues))
glue_vs_mat = list(itertools.product(glues, mats))
mat_vs_glue_paper_vs_rate = list(itertools.product(mats, glue_paper, rate))
mat_vs_glue_paper_vs_pattern_vs_cell_size = list(itertools.product(mats, glue_paper, pattern, cell_size))
mat_vs_strip_vs_pattern_vs_cell_size = list(itertools.product(mats, strip, pattern, cell_size))
mat_vs_strip_vs_pattern_vs_cell_size
```

```python
def plot_pair(df, cols, pair, xlim, ylim):
    col1, col2 = cols
    val1, val2 = pair

    ax = sb.lineplot(data=df.loc[(df[col1] == val1) & (df[col2] == val2)], x='L(mm)', y='F(N)', hue='measure', palette='flare')
    ax.set(xlim=xlim, ylim=ylim, title=f'{val1} | {val2}')

    fig = ax.get_figure()
    # fig.set_figwidth(15)
    # fig.set_figheight(10)
    fig.savefig(f'{pair}.jpg'.replace('/', '-'), dpi=400)
    fig.clf()
    return ax
```

```python
def plot_three(df, cols, pair, xlim, ylim):
    col1, col2, col3 = cols
    val1, val2, val3 = pair

    ax = sb.lineplot(data=df.loc[(df[col1] == val1) & (df[col2] == val2) & (df[col3] == val3) & (df['measure'] != 168)], x='L(mm)', y='F(N)', hue='measure', palette='flare')
    # ax.set(title=f'{val1} | {val2} | {val3}')
    ax.set(xlim=xlim, ylim=ylim, title=f'{val1} | {val2} | {val3}')

    fig = ax.get_figure()
    # fig.set_figwidth(15)
    # fig.set_figheight(10)
    fig.savefig(f'{pair}.jpg'.replace('/', '-'), dpi=400)
    fig.clf()
    return ax
```

```python
def plot_four(df, cols, pair, xlim, ylim):
    col1, col2, col3, col4 = cols
    val1, val2, val3, val4 = pair

    ax = sb.lineplot(data=df.loc[(df[col1] == val1) & (df[col2] == val2) & (df[col3] == val3) & (df[col4] == val4) & (df['measure'] != 581) & (df['measure'] != 522) & (df['measure'] != 487)], x='L(mm)', y='F(N)', hue='measure', palette='flare')
    # ax.set(title=f'{val1} | {val2} | {val3}')
    ax.set(xlim=xlim, ylim=ylim, title=f'{val1} | {val2} | {val3} | {val4}')

    fig = ax.get_figure()
    # fig.set_figwidth(15)
    # fig.set_figheight(10)
    fig.savefig(f'{pair}.jpg'.replace('/', '-'), dpi=400)
    fig.clf()
    return ax
```
